```python
from transformers import pipeline
import pandas as pd

# Sample transaction data
data = {
    "TransactionID": [1, 2, 3, 4],
    "CustomerName": ["Alice", "Bob", "Charlie", "Diana"],
    "Amount": [500, 12000, 2000, 15000],
    "Location": ["New York", "California", "Texas", "California"],
    "Flagged": [False, True, False, True]  # Flagged for fraud
}
df = pd.DataFrame(data)

# Load a Hugging Face pipeline for text generation
generator = pipeline("text-generation", model="gpt2")

# Define a message template
def generate_message(name, amount, location):
    prompt = f"Dear {name}, we detected a transaction of ${amount} in {location}. If this was not you, please contact us immediately."
    return generator(prompt, max_length=50, num_return_sequences=1)[0]["generated_text"]

# Generate messages for flagged transactions
def generate_messages(df):
    messages = []
    for _, row in df.iterrows():
        if row["Flagged"]:  # Only generate messages for flagged transactions
            message = generate_message(row["CustomerName"], row["Amount"], row["Location"])
            messages.append({"TransactionID": row["TransactionID"], "Message": message})
    return messages

# Generate and display messages
messages = generate_messages(df)
for msg in messages:
    print(f"Transaction ID: {msg['TransactionID']}\nMessage: {msg['Message']}\n")
```
